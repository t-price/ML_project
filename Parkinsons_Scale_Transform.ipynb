```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
test_X = pd.read_csv('test_X.csv')
display(test_X)
test_y = pd.read_csv('test_y.csv')
display(test_y)
train_X = pd.read_csv('train_X.csv')
display(train_X)
train_y = pd.read_csv('train_y.csv')
display(train_y)
val_X = pd.read_csv('val_X.csv')
display(val_X)
val_y = pd.read_csv('val_y.csv')
display(val_y)
```

```python
df = pd.concat([train_X, train_y], axis=1)
```

# Experimenting with Scalers and Transformers

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler, PowerTransformer

std_scaler = StandardScaler()
X_train_std = std_scaler.fit_transform(train_X)
X_val_std = std_scaler.transform(val_X)
X_test_std = std_scaler.transform(test_X)
display(X_train_std)

minmax_scaler = MinMaxScaler()
X_train_mm = minmax_scaler.fit_transform(train_X)
X_val_mm = minmax_scaler.transform(val_X)
X_test_mm = minmax_scaler.transform(test_X)
display(X_train_mm)

rob_scaler = RobustScaler()
X_train_rob = rob_scaler.fit_transform(train_X)
X_val_rob = rob_scaler.transform(val_X)
X_test_rob = rob_scaler.transform(test_X)
display(X_train_rob)

pow_transform = PowerTransformer(method='yeo-johnson')
X_train_pow = pow_transform.fit_transform(train_X)
X_val_pow = pow_transform.transform(val_X)
X_test_pow = pow_transform.transform(test_X)
display(X_train_pow)
```

```python
fig,ax = plt.subplots(2,2, figsize=(8,8))
ax[0,0].hist(X_train_std)
ax[0,0].set_title('StandardScaler')
ax[0,1].hist(X_train_mm)
ax[0,1].set_title('MinMaxScaler')
ax[1,0].hist(X_train_rob)
ax[1,0].set_title('RobustScaler')
ax[1,1].hist(X_train_pow)
ax[1,1].set_title('PowerTransformer')
plt.show()
```

# Experimenting with Isolation Forests for Anomaly Detection

```python
#use the isolation forest model to search for anomalies on the scaled/transformed data
from sklearn.ensemble import IsolationForest

# create the model
clf = IsolationForest(random_state=42, contamination=0.01)
#clf = IsolationForest(random_state=42, contamination=0.02)
anomaly_threshold = -0.5

# Fit the model to the StandardScaler data
clf.fit(X_train_std)

# create a column for the predictions of the StandardScaler data
df['anomaly_score'] = clf.decision_function(X_train_std)
df['outlier'] = clf.predict(X_train_std)
std_anomaly_test = clf.decision_function(X_test_std)
std_anomalies_test = std_anomaly_test < anomaly_threshold
std_anomaly_val = clf.decision_function(X_val_std)
std_anomalies_val = std_anomaly_val < anomaly_threshold
std_train_out = df.loc[df['outlier'] == -1].index

# show the number of outliers for the StandardScaler data
print('Standard Scaler')
print(df['outlier'].value_counts())
print(std_anomalies_test)
print(std_anomalies_val)
display(df.loc[df['outlier'] == -1])

# Fit the model to the MinMaxScaler data
clf.fit(X_train_mm)

# create a column for the predictions of the MinMaxScaler data
df['anomaly_score'] = clf.decision_function(X_train_mm)
df['outlier'] = clf.predict(X_train_mm)
mm_anomaly_test = clf.decision_function(X_test_mm)
mm_anomalies_test = mm_anomaly_test < anomaly_threshold
mm_anomaly_val = clf.decision_function(X_val_mm)
mm_anomalies_val = mm_anomaly_val < anomaly_threshold
mm_train_out = df.loc[df['outlier'] == -1].index

# show the number of outliers for the MinMaxScaler data
print('MinMaxScaler')
print(df['outlier'].value_counts())
print(mm_anomalies_test)
print(mm_anomalies_val)
display(df.loc[df['outlier'] == -1])

# Fit the model to the RobustScaler data
clf.fit(X_train_rob)

# create a column for the predictions of the RobustScaler data
df['anomaly_score'] = clf.decision_function(X_train_rob)
df['outlier'] = clf.predict(X_train_rob)
rob_anomaly_test = clf.decision_function(X_test_rob)
rob_anomalies_test = rob_anomaly_test < anomaly_threshold
rob_anomaly_val = clf.decision_function(X_val_rob)
rob_anomalies_val = rob_anomaly_val < anomaly_threshold
rob_train_out = df.loc[df['outlier'] == -1].index

# show the number of outliers for the RobustScaler data
print('RobustScaler')
print(df['outlier'].value_counts())
print(rob_anomalies_test)
print(rob_anomalies_val)
display(df.loc[df['outlier'] == -1])

# Fit the model to the PowerTransformer data
clf.fit(X_train_pow)

# create a column for the predictions for the PowerTransformer data
df['anomaly_score'] = clf.decision_function(X_train_pow)
df['outlier'] = clf.predict(X_train_pow)
pow_anomaly_test = clf.decision_function(X_test_pow)
pow_anomalies_test = pow_anomaly_test < anomaly_threshold
pow_anomaly_val = clf.decision_function(X_val_pow)
pow_anomalies_val = pow_anomaly_val < anomaly_threshold
pow_train_out = df.loc[df['outlier']==-1].index

# show the number of outliers of the PowerTransformer data
print('PowerTransformer')
print(df['outlier'].value_counts())
print(pow_anomalies_test)
print(pow_anomalies_val)
display(df.loc[df['outlier']==-1])

# Remove the outliers for each type of scaled/transformed data
X_train_no_outliers_std = np.delete(X_train_std, std_train_out, axis=0)
y_train_no_outliers_std = np.delete(train_y, std_train_out)

X_train_no_outliers_mm = np.delete(X_train_mm, mm_train_out, axis=0)
y_train_no_outliers_mm = np.delete(train_y, mm_train_out)

X_train_no_outliers_rob = np.delete(X_train_rob, rob_train_out, axis=0)
y_train_no_outliers_rob = np.delete(train_y, rob_train_out)

X_train_no_outliers_pow = np.delete(X_train_pow, pow_train_out, axis=0)
y_train_no_outliers_pow = np.delete(train_y, pow_train_out)
```

### RobustScaler SVM

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.pipeline import Pipeline
import optuna

# Define the objective function for Optuna
def objective(trial):
    # Sample hyperparameters for the SVM
    C = trial.suggest_float('C', 1e-2, 1e2, log=True)
    kernel = trial.suggest_categorical('kernel', ['linear', 'rbf', 'poly'])

    # Create the pipeline
    scaler = RobustScaler()
    svm = SVC(C=C, kernel=kernel)

    pipeline = Pipeline([
        ('scaler', scaler),
        ('svm', svm)
    ])

    # Fit the pipeline on the training data
    pipeline.fit(train_X, train_y)

    # Evaluate the pipeline on the validation set
    y_pred = pipeline.predict(test_X)
    accuracy = accuracy_score(test_y, y_pred)

    return 1.0 - accuracy  # Optuna minimizes the objective function

# Create and run the Optuna study
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=50)  # You can adjust the number of trials

# Get the best hyperparameters
best_params = study.best_params

# Train the best pipeline on the entire training set
best_pipeline = Pipeline([
    ('scaler', RobustScaler()),
    ('svm', SVC(C=best_params['C'], kernel=best_params['kernel']))
])
best_pipeline.fit(train_X, train_y)

# Evaluate the best pipeline on the test set
y_pred_best = best_pipeline.predict(test_X)
accuracy_best = accuracy_score(test_y, y_pred_best)
print(f"Pipeline Accuracy: {accuracy_best}")
```
