```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
test_X = pd.read_csv('test_X.csv')
display(test_X)
test_y = pd.read_csv('test_y.csv')
display(test_y)
train_X = pd.read_csv('train_X.csv')
display(train_X)
train_y = pd.read_csv('train_y.csv')
display(train_y)
val_X = pd.read_csv('val_X.csv')
display(val_X)
val_y = pd.read_csv('val_y.csv')
display(val_y)
```

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
X_train_std = std_scaler.fit_transform(train_X)
X_val_std = std_scaler.transform(val_X)
X_test_std = std_scaler.transform(test_X)
display(X_train_std)
```

```python
plt.hist(X_train_std)
plt.title('StandardScaler')
plt.show()
```

# Experimenting with Isolation Forests for Anomaly Detection

```python
#use the isolation forest model to search for anomalies on the scaled/transformed data
from sklearn.ensemble import IsolationForest

df = pd.concat([train_X, train_y], axis=1)

# create the model
clf = IsolationForest(random_state=42, contamination=0.01)
#clf = IsolationForest(random_state=42, contamination=0.02)
anomaly_threshold = -0.5

# Fit the model to the StandardScaler data
clf.fit(X_train_std)

# create a column for the predictions of the StandardScaler data
df['anomaly_score'] = clf.decision_function(X_train_std)
df['outlier'] = clf.predict(X_train_std)
std_anomaly_test = clf.decision_function(X_test_std)
std_anomalies_test = std_anomaly_test < anomaly_threshold
std_anomaly_val = clf.decision_function(X_val_std)
std_anomalies_val = std_anomaly_val < anomaly_threshold
std_train_out = df.loc[df['outlier'] == -1].index

# show the number of outliers for the StandardScaler data
print('Standard Scaler')
print(df['outlier'].value_counts())
print(std_anomalies_test)
print(std_anomalies_val)
display(df.loc[df['outlier'] == -1])


# Remove the outliers for each type of scaled/transformed data
X_train_no_outliers_std = np.delete(X_train_std, std_train_out, axis=0)
y_train_no_outliers_std = np.delete(train_y, std_train_out)
```

```python
fig, (ax1,ax2) = plt.subplots(1, 2, figsize=(10, 5))

clf.fit(X_train_std)
predictions_std = clf.predict(X_train_std)

ax1.scatter(X_train_std[:, 0], X_train_std[:, 1], c=clf.decision_function(X_train_std), marker='.')
ax1.set_title('StandardScaler Anomalies')
ax2.plot(X_train_std[predictions_std==1][:,0], X_train_std[predictions_std==1][:,1], 'b.')
ax2.plot(X_train_std[predictions_std==-1][:,0], X_train_std[predictions_std==-1][:,-1], 'rx')
ax2.set_title('StandardScaler Anomalies')    

df = pd.DataFrame(dict(predicted_label=clf.predict(X_train_std), score=clf.decision_function(X_train_std)))
outliers = df[df['predicted_label'] == -1].values
inliers = df[df['predicted_label'] == 1].head(len(outliers)).values
std_scores = pd.DataFrame(np.concatenate([inliers, outliers], axis=1), 
             columns=['Inlier Label', 'Inline Score', 'Outlier Label', 'Outline Score'])
display(std_scores)
      
```

## Experimenting with PCA for Dimensionality Reduction - StandardScaler

```python
#use principal component analysis to determine categories for the data
from sklearn.decomposition import PCA


#test n_number of components to find the best fit for the data
n_components = np.arange(1, 23)
models = [PCA(n).fit(X_train_std) for n in n_components]

#plot the explained variance ratio for each number of components
plt.plot(n_components, [m.explained_variance_ratio_.sum() for m in models], color='black')
for i, model in enumerate(models):
    explained_variance = model.explained_variance_ratio_.sum()
    print(f'Component {n_components[i]}: Explained Variance {explained_variance}')
plt.xlabel('n_components')
plt.ylabel('explained variance')
plt.show()


#create the model
pca = PCA(n_components = 2)

#fit the model and transform the data
transformed = pca.fit_transform(X_train_std)

#create a dataframe of the transformed data
df_ss = pd.DataFrame(transformed, columns=['x', 'y'])
df_ss['status'] = train_y

#create a scatterplot of the data
sns.scatterplot(x='x', y='y', hue='status', data=df_ss)

#make title
plt.title('PCA Scatterplot')
plt.grid()
#add lines to thicken the 0 lines
plt.axhline(0, color='black')
plt.axvline(0, color='black')

#show the plot
plt.show()
```

## Experimenting with PCA for Dimensionality Reduction - StandardScaler Outliers Removed

```python
#use principal component analysis to determine categories for the data
from sklearn.decomposition import PCA


#test n_number of components to find the best fit for the data
n_components = np.arange(1, 23)
models = [PCA(n).fit(X_train_no_outliers_std) for n in n_components]

#plot the explained variance ratio for each number of components
plt.plot(n_components, [m.explained_variance_ratio_.sum() for m in models], color='black')
for i, model in enumerate(models):
    explained_variance = model.explained_variance_ratio_.sum()
    print(f'Component {n_components[i]}: Explained Variance {explained_variance}')
plt.xlabel('n_components')
plt.ylabel('explained variance')
plt.show()


#create the model
pca = PCA(n_components = 2)

#fit the model and transform the data
transformed = pca.fit_transform(X_train_no_outliers_std)

#create a dataframe of the transformed data
df_ss_no = pd.DataFrame(transformed, columns=['x', 'y'])
df_ss_no['status'] = y_train_no_outliers_std

#create a scatterplot of the data
sns.scatterplot(x='x', y='y', hue='status', data=df_ss_no)

#make title
plt.title('PCA Scatterplot')
plt.grid()
#add lines to thicken the 0 lines
plt.axhline(0, color='black')
plt.axvline(0, color='black')

#show the plot
plt.show()
```
