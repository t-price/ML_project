```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
test_X = pd.read_csv('test_X.csv')
display(test_X)
test_y = pd.read_csv('test_y.csv')
display(test_y)
train_X = pd.read_csv('train_X.csv')
display(train_X)
train_y = pd.read_csv('train_y.csv')
display(train_y)
val_X = pd.read_csv('val_X.csv')
display(val_X)
val_y = pd.read_csv('val_y.csv')
display(val_y)

```

```python
from sklearn.preprocessing import PowerTransformer

pow_transform = PowerTransformer(method='yeo-johnson')
X_train_pow = pow_transform.fit_transform(train_X)
X_val_pow = pow_transform.transform(val_X)
X_test_pow = pow_transform.transform(test_X)
X = np.concatenate((X_train_pow, X_val_pow, X_test_pow), axis=0)
y = np.concatenate((train_y, val_y, test_y), axis=0)
display(X_train_pow)
```

```python
plt.hist(X_train_pow)
plt.title('PowerTransformer')
plt.show()
```

## Experimenting with Isolation Forests for Anomaly Detection

```python
#use the isolation forest model to search for anomalies on the scaled/transformed data
from sklearn.ensemble import IsolationForest

df = pd.concat([train_X, train_y], axis=1)

# create the model
clf = IsolationForest(random_state=42, contamination=0.01)
#clf = IsolationForest(random_state=42, contamination=0.02)
anomaly_threshold = -0.5


# Fit the model to the PowerTransformer data
clf.fit(X_train_pow)

# create a column for the predictions for the PowerTransformer data
df['anomaly_score'] = clf.decision_function(X_train_pow)
df['outlier'] = clf.predict(X_train_pow)
pow_anomaly_test = clf.decision_function(X_test_pow)
pow_anomalies_test = pow_anomaly_test < anomaly_threshold
pow_anomaly_val = clf.decision_function(X_val_pow)
pow_anomalies_val = pow_anomaly_val < anomaly_threshold
pow_train_out = df.loc[df['outlier']==-1].index

# show the number of outliers of the PowerTransformer data
print('PowerTransformer')
print(df['outlier'].value_counts())
print(pow_anomalies_test)
print(pow_anomalies_val)
display(df.loc[df['outlier']==-1])

# Remove the outliers for each type of scaled/transformed data
X_train_no_outliers_pow = np.delete(X_train_pow, pow_train_out, axis=0)
y_train_no_outliers_pow = np.delete(train_y, pow_train_out)
X_no_outliers_pow = np.delete(X, pow_train_out, axis=0)
y_no_outliers_pow = np.delete(y, pow_train_out)
```

```python
fig, (ax1,ax2) = plt.subplots(1, 2, figsize=(10, 5))

clf.fit(X_train_pow)
predictions_pow= clf.predict(X_train_pow)

ax1.scatter(X_train_pow[:, 0], X_train_pow[:, 1], c=clf.decision_function(X_train_pow), marker='.')
ax1.set_title('PowerTransformer Anomalies')
ax2.plot(X_train_pow[predictions_pow==1][:,0], X_train_pow[predictions_pow==1][:,1], 'b.')
ax2.plot(X_train_pow[predictions_pow==-1][:,0], X_train_pow[predictions_pow==-1][:,-1], 'rx')
ax2.set_title('PowerTransformer Anomalies')
plt.show()

df = pd.DataFrame(dict(predicted_label=clf.predict(X_train_pow), score=clf.decision_function(X_train_pow)))
outliers = df[df['predicted_label'] == -1].values
inliers = df[df['predicted_label'] == 1].head(len(outliers)).values
pow_scores = pd.DataFrame(np.concatenate([inliers, outliers], axis=1), 
             columns=['Inlier Label', 'Inline Score', 'Outlier Label', 'Outline Score'])
display(pow_scores)
```

### Experimenting with PCA - PowerTransformer

```python
#use principal component analysis to determine catagories for the data
from sklearn.decomposition import PCA


#test n_number of components to find the best fit for the data
n_components = np.arange(1, 23)
models = [PCA(n).fit(X_train_pow) for n in n_components]

#plot the explained variance ratio for each number of components
plt.plot(n_components, [m.explained_variance_ratio_.sum() for m in models], color='black')

best_comp_pow = None
max_expvar_pow = 0.0

for i, model in enumerate(models):
    explained_variance = model.explained_variance_ratio_.sum()
    if explained_variance > max_expvar_pow:
        max_expvar_pow = explained_variance
        best_comp_pow = n_components[i]
    print(f'Component {n_components[i]}: Explained Variance {explained_variance}')
plt.xlabel('n_components')
plt.ylabel('explained variance')
plt.show()


#create the model
pca = PCA(n_components = 8)

#fit the model and transform the data
transformed_pow = pca.fit_transform(X_train_pow)

#create a dataframe of the transformed data
df_pow = pd.DataFrame(transformed_pow, columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6','PC7','PC8'])
df_pow['status'] = train_y

#create a scatterplot of the data
sns.scatterplot(x='PC1', y='PC2', hue='status', data=df_pow)

#make title
plt.title('PCA Scatterplot')
plt.grid()
#add lines to thicken the 0 lines
plt.axhline(0, color='black')
plt.axvline(0, color='black')

#show the plot
plt.show()

#show components relations with features: 
print(pd.DataFrame(pca.components_,columns=train_X.columns,index = ['PC-1','PC-2','PC-3','PC-4','PC-5','PC-6','PC-7','PC-8']))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.pipeline import Pipeline
import optuna

# Define the objective function for Optuna
def objective(trial):
    # Sample hyperparameters for the SVM
    C = trial.suggest_float('C', 1e-2, 1e2, log=True)
    kernel = trial.suggest_categorical('kernel', ['linear', 'rbf', 'poly'])

    # Create the pipeline
    svm = SVC(C=C, kernel=kernel)

    pipeline = Pipeline([
        ('svm', svm)
    ])

    # Fit the pipeline on the training data
    pipeline.fit(X_train_pow, train_y)

    # Evaluate the pipeline on the validation set
    y_pred = pipeline.predict(val_X.values)
    accuracy = accuracy_score(val_y.values.ravel(), y_pred)
    
    return 1.0 - accuracy  # Optuna minimizes the objective function

# Create and run the Optuna study
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=50)  # You can adjust the number of trials

# Get the best hyperparameters
best_params = study.best_params

# Train the best pipeline on the entire training set
best_pipeline = Pipeline([
    ('svm', SVC(C=best_params['C'], kernel=best_params['kernel']))
])
best_pipeline.fit(X_train_pow, train_y)

# Evaluate the best pipeline on the test set
y_pred_best = best_pipeline.predict(test_X.values)
accuracy_best = accuracy_score(test_y.values.ravel(), y_pred_best)
print(f"Pipeline Accuracy: {accuracy_best}")
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.pipeline import Pipeline
import optuna

# Define the objective function for Optuna
def objective(trial):
    # Sample hyperparameters for the SVM
    C = trial.suggest_float('C', 1e-2, 1e2, log=True)
    kernel = trial.suggest_categorical('kernel', ['linear', 'rbf', 'poly'])

    # Create the pipeline
    svm = SVC(C=C, kernel=kernel)

    pipeline = Pipeline([
        ('svm', svm)
    ])

    # Fit the pipeline on the training data
    pipeline.fit(X_train_no_outliers_pow, y_train_no_outliers_pow)

    # Evaluate the pipeline on the validation set
    y_pred = pipeline.predict(val_X.values)
    accuracy = accuracy_score(val_y.values.ravel(), y_pred)
    
    return 1.0 - accuracy  # Optuna minimizes the objective function

# Create and run the Optuna study
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=50)  # You can adjust the number of trials

# Get the best hyperparameters
best_params = study.best_params

# Train the best pipeline on the entire training set
best_pipeline = Pipeline([
    ('svm', SVC(C=best_params['C'], kernel=best_params['kernel']))
])
best_pipeline.fit(X_train_no_outliers_pow, y_train_no_outliers_pow)

# Evaluate the best pipeline on the test set
y_pred_best = best_pipeline.predict(test_X.values)
accuracy_best = accuracy_score(test_y.values.ravel(), y_pred_best)
print(f"Pipeline Accuracy: {accuracy_best}")
```
